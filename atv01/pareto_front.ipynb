```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def no_dominated_min_min(pontos):
  """"
  Determina os pontos não dominados de Pareto
    
  Input: 
  pontos = conjunto de pontos [f1 f2]

  Output:
  nd_pontos = pontos não dominados [f1* f2*]

  """

  N = len(pontos['0']) # quantidade  de pontos
  nd_pontos = []
    
  while N!=0:
    # Há pontos para classificar.
    nd = True

    for i in range(len(pontos['0'])):
      if i == N - 1:
        continue

      if not((pontos.iat[N - 1, 1] == pontos.iat[i, 1]) and (pontos.iat[N - 1, 2] == pontos.iat[i, 2])):
        if ((pontos.iat[i, 1] < pontos.iat[N - 1, 1]) and (pontos.iat[i, 2] <= pontos.iat[N - 1, 2])) or ((pontos.iat[i, 1] <= pontos.iat[N - 1, 1]) and (pontos.iat[i, 2] < pontos.iat[N - 1, 2])):
          nd = False
          break
    
    if nd:
      nd_pontos.append(N - 1)
    N -= 1

    #None # encontre os pontos não dominados
    
    
  return nd_pontos # retorne os pontos não dominados
```

```python
# Min F1 - Min F2
x = no_dominated_min_min(pontos)
nd_points_x_axis = []
nd_points_y_axis = []
d_points_x_axis = []
d_points_y_axis = []

for i in range(len(pontos['0'])):
    if i in x:
        nd_points_x_axis.append(pontos.iat[i, 1])
        nd_points_y_axis.append(pontos.iat[i, 2])
    else:
        d_points_x_axis.append(pontos.iat[i, 1])
        d_points_y_axis.append(pontos.iat[i, 2])


x_axis = np.array(d_points_x_axis)
y_axis = np.array(d_points_y_axis)
plt.scatter(x_axis, y_axis, color = 'black')

x_axis = np.array(nd_points_x_axis)
y_axis = np.array(nd_points_y_axis)
plt.scatter(x_axis, y_axis, color = 'blue')
plt.title("Minimize F1 and Minimize F2")
plt.show()
```
